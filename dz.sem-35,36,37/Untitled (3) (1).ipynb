### ЗАДАНИЕ 35. Создайте двусвязный список для хранения информации о задачах в календаре. Каждый элемент списка должен содержать название задачи, описание, дату начала и дату окончания, а также список напоминаний о задаче (например, за день до окончания, за час до начала и т.д.).

```python
class Task:
    def __init__(self, name, description, start_date, end_date): #класс задач, содержит атрибуты имя, описание, время начала
#дата окончания и напоминания
        self.name = name
        self.description = description
        self.start_date = start_date
        self.end_date = end_date
        self.reminders = []

class Reminder: #класс напоминание, содержит атрибуты время до события и сообщение напоминания
    def __init__(self, time_before, message):
        self.time_before = time_before
        self.message = message

class Node: #узел двусвязного списка, содержит атрибуты задач, ссылку на предыдущий и следующий узел
    def __init__(self, task):
        self.task = task
        self.prev = None
        self.next = None

class DoublyLinkedList: #сам двусвязный список задач
#внутри класса атрибут head (головной узел) и методы для вставки задачи в список (insert_task)
#и отображения всех задач в списке (display_tasks)
    def __init__(self):
        self.head = None

    def insert_task(self, task):
        new_node = Node(task)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def display_tasks(self):
        current = self.head
        while current:
            print("Задание:", current.task.name)
            print("Описание:", current.task.description)
            print("Начало:", current.task.start_date)
            print("Конец:", current.task.end_date)
            print("Напоминания:")
            for reminder in current.task.reminders:
                print("- Времени до:", reminder.time_before)
                print("- Сообщение:", reminder.message)
            print()
            current = current.next

#создаем список задач и добавляем напоминания
task1 = Task("Задание 1", "Доделать отчет по проекту", "2052-05-07", "2052-05-12")
task1.reminders.append(Reminder("1 день", "Предоставить отчет завтра"))
task1.reminders.append(Reminder("1 час", "Закончите отчет через час"))

task2 = Task("Задание 2", "Подготовить презентацию", "2052-05-18", "2052-05-23")
task2.reminders.append(Reminder("3 дня", "Попрактиковать выступление"))

task3 = Task("Задание 3", "Посетить конференцию", "2052-05-25", "2052-05-28")
task3.reminders.append(Reminder("1 день", "Получить билет на конференцию"))

task_list = DoublyLinkedList()
task_list.insert_task(task1)
task_list.insert_task(task2)
task_list.insert_task(task3)

task_list.display_tasks()
```

### ЗАДАНИЕ 36. Реализовать функцию, которая находит средний элемент в двусвязном списке.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def find_middle_element(self):
        if self.head is None:
            return None

        slow_ptr = self.head
        fast_ptr = self.head

        while fast_ptr is not None and fast_ptr.next is not None:
            slow_ptr = slow_ptr.next
            fast_ptr = fast_ptr.next.next

        return slow_ptr.data

#создаем список и добавляем элементы
linked_list = DoublyLinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
linked_list.append(4)
linked_list.append(5)

middle_element = linked_list.find_middle_element()
print("Средний элемент в списке:", middle_element)
```

### ЗАДАНИЕ 37. Реализовать функцию, которая сортирует циклический двусвязный список по возрастанию.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.prev = self.head
            self.head.next = self.head
        else:
            last_node = self.head.prev
            last_node.next = new_node
            new_node.prev = last_node
            new_node.next = self.head
            self.head.prev = new_node

    def sort(self):
        if not self.head:
            return
        current = self.head
        while True:
            if current.next == self.head:
                break
            if current.data > current.next.data:
                temp = current.data
                current.data = current.next.data
                current.next.data = temp
                current = self.head
            else:
                current = current.next

    def display(self):
        if not self.head:
            return
        current = self.head
        while True:
            print(current.data)
            current = current.next
            if current == self.head:
                break

# Пример использования
dll = DoublyLinkedList()
dll.append(3)
dll.append(1)
dll.append(4)
dll.sort()
dll.display()
```
