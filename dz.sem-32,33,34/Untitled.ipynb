### ЗАДАЧА 32
Используя модуль array и библиотеку numpy, решить следующие задачи:
а) Дан массив размера  N . Найти номера двух ближайших элементов из этого массива (т. е. элементов с наименьшим модулем разности) и вывести эти номера в порядке возрастания.
б) Дана целочисленная матрица размера  M×N , элементы которой могут принимать значения от 0 до 100. Различные строки матрицы назовем похожими, если совпадают множества чисел, встречающихся в этих строках. Найти количество строк, похожих на первую строку данной матрицы.

```python
#ПУНКТ А
import numpy as np
from array import array

# Заданный массив
arr = array('i', [10, 99, 7, 20, 13, 25, 12])

# Индексы двух ближайших элементов
indices = np.argsort(np.abs(np.diff(arr)))[:2]

# Вывод номеров в порядке возрастания
sorted_indices = np.sort(indices)
print(f"Номера двух ближайших элементов: {sorted_indices}")
```

```python
#ПУНКТ Б
import numpy as np

# Запрашиваем у пользователя размеры матрицы
M = int(input("Введите количество строк матрицы: "))
N = int(input("Введите количество столбцов матрицы: "))

# Создаем пустую матрицу
matrix = np.zeros((M, N), dtype=int)

# Заполняем матрицу вручную
for i in range(M):
    for j in range(N):
        matrix[i][j] = int(input(f"Введите элемент [{i}][{j}]: "))

print("Исходная матрица:")
print(matrix)

# Находим множество чисел в первой строке
first_row_set = set(matrix[0])

# Считаем количество строк, похожих на первую строку
similar_rows_count = 0
for i in range(1, M):
    if set(matrix[i]) == first_row_set:
        similar_rows_count += 1

print(f"Количество строк, похожих на первую строку: {similar_rows_count}")
```

```python
class Stack(list):
    def push(self, e):
        self.append(e)
    def is_empty(self):
        return len(self) == 0
    def __pop__(self):
        return self.pop(-1)
    def peek(self):
        return self[-1]
    def size(self):
        return len(self)
st = Stack([1,2,3,4,5,6,7])
st1 = Stack()
A = 2
B = 4
print(st)
st.pop()
print(st)
for i in range(st.size()):
    el = st.peek()
    if el == A or el == B:
        st.pop()
        el = st.peek()
        while el != A and el != B:
            st.pop()
            el = st.peek()
        st.pop()
        break
    else:
        st1.push(st.pop())
for el in st1[::-1]:st.push(el)
st
```

ВАРИАНТ 5 - Создать класс очереди, который будет поддерживать операции добавления элемента в конец очереди, удаления элемента из начала очереди и сортировки элементов в очереди по возрастанию.

```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None

    def sort_queue(self):
        self.items.sort()

    def is_empty(self):
        return len(self.items) == 0

# Пример использования класса Queue
q = Queue()
q.enqueue(3)
q.enqueue(1)
q.enqueue(2)

print("Очередь до сортировки:", q.items)

q.sort_queue()
print("Очередь после сортировки:", q.items)

print("Удален элемент из начала очереди:", q.dequeue())
print("Очередь после удаления элемента:", q.items)
q.enqueue(88)
print(q.items) #добавление элемента в конец
```
